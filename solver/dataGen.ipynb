### N = 10

    N = 10
    M = 2
    t = [2, 4]
    g = 2
    s = [25, 40]
    c = [30, 40]

```python
import random as rd
rd.seed(0)

N = 10
M = 2
t = []
g = []
s = []
c = [30, 40]

def dataGen1():
  print(N, M)
  print(5, 12)
  for i in range(N):
    t.append(rd.randint(2, 4))
    g.append(rd.randint(1, 2))
    s.append(rd.randint(25, 40))
    print(str(t[i]) + ' ' + str(g[i]) + ' ' + str(s[i]))
  for i in range(M):
    print(c[i], end=' ')

import sys
original_stdout = sys.stdout

with open('/content/drive/MyDrive/HUST/20202/Tối ưu lập kế hoạch/data/test1.txt', 'w') as f:
  sys.stdout = f
  dataGen1()
  sys.stdout = original_stdout
```

## Final

```python
def dataGen(name='test.txt', N=10, M=2, U=5, K=12, T=[2,4], G=10, S=[25,40], C=[30, 40]):
  import random as rd
  import sys
  rd.seed(0)
  original_stdout = sys.stdout
  file_name = '/content/drive/MyDrive/HUST/20202/Tối ưu lập kế hoạch/data/' + name
  with open(file_name, 'w') as f:
    sys.stdout = f

    t = []
    g = []
    s = []
    c = []
    print(N, M)
    print(U, K)
    for i in range(N):
      t.append(rd.randint(T[0], T[1]))
      g.append(rd.randint(1, G))
      s.append(rd.randint(S[0], S[1]))
      print(str(t[i]) + ' ' + str(g[i]) + ' ' + str(s[i]))
    for i in range(M):
      c.append(rd.choice([C[0], C[1]]))
      print(c[i], end=' ')

    sys.stdout = original_stdout
```
